```python
from collections import defaultdict

# Sample documents (you can replace these with your own)
documents = {
    1: "machine learning provides systems the ability to automatically learn",
    2: "deep learning is a part of machine learning",
    3: "information retrieval deals with search and data mining",
}

# Step 1: Create an inverted index
inverted_index = defaultdict(list)

for doc_id, text in documents.items():
    words = text.lower().split()
    for word in set(words):  # use set() to avoid duplicate entries in same doc
        inverted_index[word].append(doc_id)

# Step 2: Display inverted index
print("ğŸ” Inverted Index:\n")
for word, doc_list in inverted_index.items():
    print(f"{word} â†’ {doc_list}")

# Step 3: Document retrieval based on query
def retrieve(query):
    query_words = query.lower().split()
    result_docs = set()
    
    for word in query_words:
        if word in inverted_index:
            if not result_docs:
                result_docs = set(inverted_index[word])
            else:
                result_docs &= set(inverted_index[word])  # intersection for AND query
        else:
            result_docs = set()
            break
    return result_docs

# Step 4: Example query
query = "machine learning"
results = retrieve(query)

print("\nğŸ” Query:", query)
if results:
    print("Documents Retrieved:", results)
    for doc_id in results:
        print(f"Doc {doc_id}: {documents[doc_id]}")
else:
    print("No matching documents found.")
```
